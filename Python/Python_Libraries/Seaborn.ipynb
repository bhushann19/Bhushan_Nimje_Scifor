## 1. Introduction to Seaborn

Seaborn is a powerful and flexible data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn is particularly well-suited for visualizing complex datasets and creating elegant plots with minimal code. 

To use Seaborn, you first need to import it. It is often imported alongside Matplotlib for additional customization.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Built-in Datasets

Seaborn comes with several built-in datasets for easy visualization practice.

```python
# Loading a built-in dataset
tips = sns.load_dataset('tips')
print(tips.head())
```

## 3. Scatter Plots

### Basic Scatter Plot

```python
# Creating a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

###  Hue

The hue parameter is used to add a semantic variable to the plot, typically by using different colors to distinguish between categories or groups. This allows for better visualization and understanding of relationships within the data.

How hue Works:-

The hue parameter is used in many Seaborn functions to create a color-coded distinction between different levels of a categorical variable. This can be particularly useful in plots such as scatter plots, line plots, bar plots, etc.

### Scatter Plot with Hue

```python
# Scatter plot with hue for categorical separation
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip by Time')
plt.show()
```

## 4. Line Plots

```python
# Loading a built-in dataset
fmri = sns.load_dataset('fmri')

# Creating a line plot
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.title('Line Plot of Signal over Time')
plt.show()
```

### Line Plot with Hue

```python
# Line plot with hue for categorical separation
sns.lineplot(x='timepoint', y='signal', hue='event', data=fmri)
plt.title('Line Plot of Signal over Time by Event')
plt.show()
```

## 5. Bar Plots

```python
# Creating a bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

### Bar Plot with Hue

```python
# Bar plot with hue for categorical separation
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title('Bar Plot of Total Bill by Day and Sex')
plt.show()
```

## 6. Box Plots

```python
# Creating a box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

### Box Plot with Hue

```python
# Box plot with hue for categorical separation
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.title('Box Plot of Total Bill by Day and Smoker Status')
plt.show()
```

## 7. Violin Plots

```python
# Creating a violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

### Violin Plot with Hue

```python
# Violin plot with hue for categorical separation
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, data=tips)
plt.title('Violin Plot of Total Bill by Day and Sex')
plt.show()
```

## 8. Heatmaps

```python
# Creating a correlation matrix
corr = tips.corr()

# Creating a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Tips Dataset Correlation Matrix')
plt.show()
```

## 9. Customizing Seaborn Plots

### Setting Style and Context

```python
# Setting the style
sns.set_style('whitegrid')

# Setting the context
sns.set_context('talk')

# Creating a customized plot
sns.scatterplot(x='total_bill', y='tip', hue='time', style='sex', data=tips)
plt.title('Customized Scatter Plot of Total Bill vs Tip')
plt.show()
```

### Custom Color Palettes

```python
# Custom color palette
palette = sns.color_palette("husl", 8)

# Creating a plot with a custom color palette
sns.boxplot(x='day', y='total_bill', palette=palette, data=tips)
plt.title('Box Plot of Total Bill by Day with Custom Palette')
plt.show()
```
