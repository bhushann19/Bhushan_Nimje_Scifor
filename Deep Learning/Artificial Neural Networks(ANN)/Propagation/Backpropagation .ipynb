Backward Propagation, or backpropagation, is the process of adjusting the weights in a neural network to minimize the loss function. It is a key part of the training phase in neural networks, involving the following steps:

1. **Forward Pass:** Calculate the output of the network using the current weights.
2. **Compute Loss:** Measure the difference between the predicted output and the actual target using a loss function.
3. **Backward Pass:** Compute the gradient of the loss function with respect to each weight using the chain rule of calculus.
4. **Update Weights:** Adjust the weights to minimize the loss, typically using gradient descent.
