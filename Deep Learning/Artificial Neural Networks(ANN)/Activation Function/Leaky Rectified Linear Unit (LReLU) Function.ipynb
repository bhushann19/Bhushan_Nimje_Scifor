1. The Leaky Rectified Linear Unit (LReLU) activation function is a variant of the standard Rectified Linear Unit (ReLU) activation function. It's designed to address the "dying ReLU" problem, where ReLU neurons can become inactive during training, effectively outputting zero for all inputs. LReLU allows a small, non-zero gradient when the input is negative, preventing neurons from dying completely.The mathematical definition of LReLU is:

    LReLU(x)= $\begin{cases}0.01 x & \text { if } x<0 \\ x & \text { if } x \geq 0\end{cases}$

2. It is slightly modified version of ReLU function & it is also commonly used.
3. Rather than completely giving 0.0 as an output for the negative input it slightly leaks the output.



### Properties of Leaky ReLU Activation Function:

1. **Output Range:** Like ReLU, LReLU returns the input x if it's positive. However, for negative inputs, it returns x, where  is a small positive slope (typically a small value like 0.01).

2. **Avoids Dying ReLU Problem:** By allowing a small, non-zero gradient for negative inputs, LReLU prevents neurons from becoming completely inactive during training, addressing the "dying ReLU" problem.

3. **Non-Linearity:** LReLU introduces non-linearity into the network, allowing it to model complex relationships in the data, similar to ReLU.

4. **Computationally Efficient:** LReLU is computationally efficient to compute, similar to ReLU, since it involves only simple thresholding operations.

### Advantages of Leaky ReLU Activation Function:

1. **Prevents Neurons from Dying:** LReLU prevents neurons from becoming completely inactive during training by allowing a small, non-zero gradient for negative inputs. This helps in mitigating the "dying ReLU" problem.

2. **Faster Convergence:** LReLU tends to converge faster during training compared to traditional ReLU, especially in situations where neurons would otherwise die out.

3. **Sparse Activation:** LReLU can still introduce sparsity in activations by setting negative values to a small fraction of their input. This can help in reducing overfitting and making computations more efficient.

### Drawbacks of Leaky ReLU Activation Function:

**Tuning the Slope Parameter:** The slope parameter  needs to be tuned empirically, which adds an additional hyperparameter to the model.

### Usage of Leaky ReLU Activation Function:

LReLU activation function is commonly used in neural networks, especially in situations where the standard ReLU activation function leads to dying neurons. It's often preferred over ReLU in deep learning architectures like convolutional neural networks (CNNs) and feedforward neural networks (FNNs) due to its ability to prevent neurons from becoming completely inactive during training.

Overall, Leaky ReLU activation function is a powerful variant of the standard ReLU activation function, offering advantages such as preventing neurons from dying out and faster convergence during training. However, it's important to tune the slope parameter  appropriately based on the specific requirements of the task at hand.
