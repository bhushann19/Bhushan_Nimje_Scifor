1. The Hyperbolic Tangent (Tanh) activation function is another popular non-linear activation function used in neural networks. It's defined mathematically as:

$\tanh (x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}$

2. It is very similar to sigmoid function and also has the same S-shaped curve with the difference in the output range of -1 to 1.
3. In tanh if the input is large then output will be very close to +1.0 and if the input is smaller then the output will be near to -1.0.



### Properties of Tanh Activation Function:

1. **Output Range:** The tanh function squashes the input values to the range between -1 and 1, similar to the sigmoid function but zero-centered. This property is particularly useful in situations where the mean of the input data is not centered around zero.

2. **Smoothness:** Like the sigmoid function, the tanh function is smooth and continuously differentiable, making it suitable for optimization algorithms that rely on gradient descent-based methods.

3. **Zero-Centered:** One of the key advantages of the tanh function over the sigmoid function is that its output is zero-centered. This property can help mitigate issues like the "zigzagging" of weights during training, especially in deep networks.

4. **Non-Linearity:** Similar to other activation functions, tanh introduces non-linearity into the network, allowing it to model complex relationships in the data.

### Advantages of Tanh Activation Function:

1. **Zero-Centered:** The tanh function has a zero-centered output, which can help in faster convergence during training, especially in deep networks.

2. **Output Range:** The output of the tanh function is bounded between -1 and 1, which can help prevent issues like the exploding gradient problem.

3. **Smooth Gradient:** Tanh has a smooth gradient across most of its domain, aiding in stable and consistent learning during training.

### Drawbacks of Tanh Activation Function:

**Vanishing Gradient:** Like the sigmoid function, tanh also suffers from the vanishing gradient problem, especially for very large or very small input values. This can lead to slow convergence during training, particularly in deep networks.

### Usage of Tanh Activation Function:

Tanh activation function is commonly used in hidden layers of neural networks, especially in recurrent neural networks (RNNs) and long short-term memory networks (LSTMs). Its zero-centered output makes it a preferred choice over the sigmoid function, particularly in situations where the mean of the input data is not centered around zero. However, similar to other activation functions, it's important to choose tanh carefully based on the characteristics of the problem and the architecture of the neural network.

Overall, tanh activation function is a versatile and widely used activation function in neural networks, offering advantages such as zero-centered output and bounded output range. However, it's important to be aware of its limitations, such as the vanishing gradient problem, and choose activation functions accordingly based on the specific requirements of the task at hand.
