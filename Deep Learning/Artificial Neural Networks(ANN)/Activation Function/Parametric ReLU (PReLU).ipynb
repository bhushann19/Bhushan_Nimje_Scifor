1. The Parametric Rectified Linear Unit (PReLU) is an extension of the standard ReLU and Leaky ReLU activation functions. It introduces a learnable parameter that controls the slope of the negative part of the function, providing greater flexibility and potentially better performance. The mathematical definition of PReLU is:

    PReLU(x) = $\begin{cases}\alpha x & \text { if } x<0 \\ x & \text { if } x \geq 0\end{cases}$
    Here, α is not a fixed small constant as in Leaky ReLU, but a learnable parameter that gets updated during the training process through backpropagation.
    
2. Only limitation of this function is that it may perform differently for different problems depending upon the value of slope parameter a.



### Properties of PReLU Activation Function:

1. **Output Range:** PReLU returns the input x if it's positive. For negative inputs, it returns αx, where α is a parameter learned during training.

2. **Learnable Slope:** The key feature of PReLU is the learnable slope α for negative inputs. This allows the network to adaptively learn the appropriate slope, potentially improving performance.

3. **Non-Linearity:** PReLU introduces non-linearity into the network, enabling it to model complex relationships in the data, similar to ReLU and Leaky ReLU.

4. **Computationally Efficient:** PReLU is computationally efficient, similar to ReLU and Leaky ReLU, involving only simple thresholding operations.

### Advantages of PReLU Activation Function:

1. **Adaptive Learning:** PReLU allows the network to adaptively learn the slope of the activation function for negative inputs, potentially leading to better performance compared to fixed-slope functions like ReLU and Leaky ReLU.

2. **Prevents Neurons from Dying:** By learning the slope parameter α, PReLU can effectively prevent neurons from becoming inactive during training, mitigating the "dying ReLU" problem.

3. **Improved Flexibility:** The flexibility to learn the slope parameter can result in improved convergence and potentially better model performance on a wide range of tasks.

### Drawbacks of PReLU Activation Function:

1. **Additional Parameters:** PReLU introduces additional parameters that need to be learned during training. This can slightly increase the complexity of the model and the computational cost.

2. **Risk of Overfitting:** With the added flexibility of learnable parameters, there is a potential risk of overfitting, especially if the dataset is small.

### Usage of PReLU Activation Function:

PReLU activation function is commonly used in neural networks, particularly in deep learning architectures like convolutional neural networks (CNNs) and feedforward neural networks (FNNs). It is often preferred when the standard ReLU or Leaky ReLU functions do not perform optimally, or when there is a need for the model to adaptively learn the activation characteristics.
