## What is a Cost Function?

It is a function that measures the performance of a model for any given data. Cost Function quantifies the error between predicted values and expected values and presents it in the form of a single real number.

## Gradient Descent (GD) Optimizer

Gradient Descent is a fundamental optimization algorithm used for minimizing the loss function in neural networks. The basic idea is to iteratively adjust the network's parameters (weights and biases) in the direction of the negative gradient of the loss function with respect to the parameters. This process continues until the loss function converges to a minimum.



### How Gradient Descent Works


**1. Initialization:** Start with an initial guess for the parameters (often random values).

**2. Compute Gradient:** Calculate the gradient of the loss function with respect to each parameter.

**3. Update Parameters:** Adjust the parameters in the direction of the negative gradient.

**4. Iterate:** Repeat the process until the parameters converge to values that minimize the loss function.

### Types of Gradient Descent Optimizers

1. Batch Gradient Descent (BGD)
2. Stochastic Gradient Descent (SGD)
3. Mini-Batch Gradient Descent

## 1. Batch Gradient Descent (BGD)

### How It Works:

1. Uses the entire training dataset to compute the gradient of the loss function.
2. Updates the parameters after processing all training examples.

### Advantages:

1. Provides a stable convergence because it uses the entire dataset.
2. Gradient is accurate and less noisy.

### Disadvantages:

1. Can be very slow and computationally expensive for large datasets.
2. Requires a lot of memory as it processes the entire dataset at once.

## 2. Stochastic Gradient Descent (SGD)

### How It Works:

1. Uses one training example at a time to compute the gradient and update the parameters.
2. Parameters are updated after each training example.

### Advantages:

1. Faster convergence compared to batch gradient descent for large datasets.
2. Introduces noise in the gradient updates, which can help escape local minima.

### Disadvantages:

1. The updates can be noisy and cause the loss function to fluctuate rather than smoothly converge.
2. Requires more iterations to converge compared to batch gradient descent.

## 3. Mini-Batch Gradient Descent

### How It Works:

1. Divides the training dataset into small batches and uses each batch to compute the gradient.
2. Updates the parameters after processing each batch.

### Advantages:

1. Provides a balance between the computational efficiency of SGD and the stability of BGD.
2. Allows for more efficient use of computing resources (e.g., GPU) compared to SGD.

### Disadvantages:

1. Requires careful tuning of the batch size.
2. Gradient updates can still be noisy, but less so than SGD.
