Probability distributions describe how the values of a random variable are distributed. Here, we will discuss three key types of probability distributions:

### 1. Normal Distribution

The normal distribution is defined by its mean and standard deviation. It is symmetric about the mean and follows a bell-shaped curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data for the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

title = "Normal Distribution\nMean = {:.2f}, Std Dev = {:.2f}".format(mean, std_dev)
plt.title(title)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Normal Distribution:

* We use np.random.normal(loc=mean, scale=std_dev, size=1000) to generate 1000 data points following a normal distribution with a specified mean and standard deviation.
* plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black') plots the histogram of the data.
* We use norm.pdf(x, mean, std_dev) from SciPy to plot the probability density function (PDF) of the normal distribution.

### 2. Binomial Distribution

The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials (experiments), each with the same probability of success.

```python
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate data for the binomial distribution
data = np.random.binomial(n, p, 1000)

# Plot the histogram of the data
plt.hist(data, bins=n+1, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the binomial distribution probabilities
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'k', marker='o', linestyle='None')

title = "Binomial Distribution\nn = {}, p = {:.2f}".format(n, p)
plt.title(title)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

Binomial Distribution:

* We use np.random.binomial(n, p, 1000) to generate 1000 data points following a binomial distribution with n trials and probability of success p.
* plt.hist(data, bins=n+1, density=True, alpha=0.6, color='b', edgecolor='black') plots the histogram of the data.
* We use binom.pmf(x, n, p) from SciPy to plot the probability mass function (PMF) of the binomial distribution.

### 3. Poisson Distribution

The Poisson distribution describes the number of events occurring within a fixed interval of time or space, given the average number of occurrences.

```python
from scipy.stats import poisson

# Parameter for the Poisson distribution
lambda_ = 5  # Average number of events per interval

# Generate data for the Poisson distribution
data = np.random.poisson(lambda_, 1000)

# Plot the histogram of the data
plt.hist(data, bins=np.max(data)-np.min(data), density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the Poisson distribution probabilities
x = np.arange(0, np.max(data))
pmf = poisson.pmf(x, lambda_)
plt.plot(x, pmf, 'k', marker='o', linestyle='None')

title = "Poisson Distribution\nλ = {:.2f}".format(lambda_)
plt.title(title)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

Poisson Distribution:

* We use np.random.poisson(lambda_, 1000) to generate 1000 data points following a Poisson distribution with average rate lambda_.
* plt.hist(data, bins=np.max(data)-np.min(data), density=True, alpha=0.6, color='b', edgecolor='black') plots the histogram of the data.
* We use poisson.pmf(x, lambda_) from SciPy to plot the probability mass function (PMF) of the Poisson distribution.

### 4. Cumulative Distribution Function (CDF)

The Cumulative Distribution Function (CDF) of a random variable X describes the probability that X will take a value less than or equal to x. Formally, for a continuous random variable, the CDF (x)
F(x) is defined as:
F(x) = P(X<=x)

The CDF is a non-decreasing function that ranges from 0 to 1. It provides a complete description of the distribution of a random variable.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data for the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Sort the data
data_sorted = np.sort(data)

# Calculate the CDF values
cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

# Plot the CDF
plt.plot(data_sorted, cdf, marker='.', linestyle='none')
plt.title("CDF of Normal Distribution\nMean = {:.2f}, Std Dev = {:.2f}".format(mean, std_dev))
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

Normal Distribution:

* We generate 1000 random data points from a normal distribution using np.random.normal(loc=mean, scale=std_dev, size=1000).
* We sort the data using np.sort(data).
* We calculate the CDF values as the cumulative count of sorted data points divided by the total number of points using np.arange(1, len(data_sorted) + 1) / len(data_sorted).
* We plot the sorted data against their corresponding CDF values.

### 5. Symmetric Distribution

A symmetric distribution is a type of probability distribution where the left and right sides of the distribution are mirror images of each other. This means that for any given point x on the distribution, the probability density at x is the same as at (−x). The most common example of a symmetric distribution is the normal distribution.

Characteristics of Symmetric Distributions:-

* Mean, Median, and Mode: In a perfectly symmetric distribution, the mean, median, and mode are all equal and located at the center of the distribution.
* Shape: The left and right tails of the distribution are mirror images.
* Examples: Besides the normal distribution, other examples include the uniform distribution (if centered around its mean) and the t-distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data for the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

title = "Symmetric Normal Distribution\nMean = {:.2f}, Std Dev = {:.2f}".format(mean, std_dev)
plt.title(title)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 6. Skewed Right Distribution

A right-skewed (or positively skewed) distribution is a type of distribution where the tail on the right side is longer or fatter than the left side. In other words, the bulk of the values lie to the left of the mean, with a few larger values stretching the tail to the right. This results in the mean being greater than the median, which is often greater than the mode.

Characteristics of Right-Skewed Distributions:-

* Mean, Median, Mode: In a right-skewed distribution, the mean is typically greater than the median, which is greater than the mode.
* Shape: The right tail (higher values) is longer and more stretched out compared to the left tail.
* Examples: Common examples of right-skewed distributions include income distribution, housing prices, and certain biological measurements.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the exponential distribution
scale = 2.0  # This is the inverse of the rate parameter (λ = 1/scale)

# Generate data for the exponential distribution
data = np.random.exponential(scale, 1000)

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='r', edgecolor='black')

# Plot the exponential distribution curve
x = np.linspace(0, np.max(data), 1000)
pdf = (1/scale) * np.exp(-x/scale)
plt.plot(x, pdf, 'k', linewidth=2)

title = "Right-Skewed Exponential Distribution\nScale = {:.2f}".format(scale)
plt.title(title)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 7. Skewed Left Distribution

A left-skewed (or negatively skewed) distribution is a type of distribution where the tail on the left side is longer or fatter than the right side. In other words, the bulk of the values lie to the right of the mean, with a few smaller values stretching the tail to the left. This results in the mean being less than the median, which is often less than the mode.

Characteristics of Left-Skewed Distributions:-

* Mean, Median, Mode: In a left-skewed distribution, the mean is typically less than the median, which is less than the mode.
* Shape: The left tail (lower values) is longer and more stretched out compared to the right tail.
* Examples: Common examples of left-skewed distributions include age at retirement, scores on an easy exam, and certain financial metrics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data for the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Transform the data to make it left-skewed
left_skewed_data = -np.exp(data)

# Plot the histogram of the left-skewed data
plt.hist(left_skewed_data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Add a title and labels
plt.title("Left-Skewed Distribution")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### 8. Central Limit Theorem (CLT) 

The Central Limit Theorem (CLT) is a fundamental statistical principle that states that the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables, regardless of the original distribution of the population, will approximate a normal distribution. This approximation improves with larger sample sizes.

Key Points of the Central Limit Theorem:-

* Independence: The sampled observations must be independent. This means the outcome of one observation should not influence the outcome of another.
* Identically Distributed: The observations must come from the same distribution with a finite mean (μ) and variance (σ²).
* Sample Size: The sample size should be sufficiently large. While there's no strict rule, a common threshold is n ≥ 30.
* Normal Approximation: As the sample size increases, the distribution of the sample mean tends to be normally distributed with mean μ (the population mean) and standard deviation σ/√n (the population standard deviation divided by the square root of the sample size).
