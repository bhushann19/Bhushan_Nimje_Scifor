```python
story = """
In the rugged terrain of the Western Ghats, Shivaji Maharaj, the founder of the Maratha Empire, was preparing for another day of strategic planning and military campaigns. Known for his innovative tactics and leadership, Shivaji Maharaj was determined to strengthen his kingdom and protect his people from external threats.
Shivaji Maharaj’s stronghold was the Raigad Fort, a formidable fortress perched high on a hilltop. The fort's strategic location allowed him to oversee the surrounding lands and defend against invaders. His trusted advisors and generals would gather in the fort's grand hall to discuss plans, share intelligence, and make critical decisions for the empire's expansion.
The evenings in Shivaji Maharaj's court were vibrant with activity. He would host feasts and gatherings to bolster the morale of his soldiers and celebrate victories. During these events, Shivaji Maharaj would address his people, sharing his vision for a prosperous and united Maratha Empire. His leadership inspired loyalty and courage among his subjects.
As night fell, Shivaji Maharaj would retire to his quarters, reflecting on the day's achievements and preparing for the challenges ahead. His dreams were filled with visions of a strong and resilient empire, where justice and valor would prevail.
"""
```

### 1. Tokenization

```python
import nltk 
from nltk.tokenize import sent_tokenize, word_tokenize

# a. Tokenize the story into sentences.
sentences = sent_tokenize(story)
print("\nTokenized Sentences: ")
print(sentences)

print('\n')
print("=" * 50)
print('\n')

# b. Tokenize the story into words.
words = word_tokenize(story)
print("Tokenized Words: ")
print(words)
```

### Lowercasing

```python
story_lowercase = story.lower()
print(story_lowercase)
```

### Stopword Removal

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
removed_stopwords = [word for word in words if word.lower() in stop_words]
print("Story without Stopwords: ")
print(' '.join(filtered_words))
print("\nRemoved Stopword: ")
print(removed_stopwords)
```

### Punctuation Removal

```python
import string

cleaned_text = story.translate(str.maketrans('','', string.punctuation))
print("\nText without Punctuation:")
print(cleaned_text)
```

### Stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
print("\nStemmed Words:")
print(stemmed_words[:10])
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmitizer = WordNetLemmatizer()
lemmitized_words = [lemmitizer.lemmatize(word) for word in filtered_words]
print("\nLemmitized Words:")
print(lemmitized_words[:10])
```

### Named Entity Recognition (NER)

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

def ner(text):                    
  #tokenize the text
  word_tokens = word_tokenize(text)

  #pos tagging of words
  word_pos = pos_tag(word_tokens)

  #tree of word entities
  print(ne_chunk(word_pos))

ner(story)
```

### Part-of-Speech Tagging

```python
from textblob import TextBlob

pos_tagging = TextBlob(story)
pos_tagging.tags
```

### Term Frequency

```python
from collections import Counter

term_freq = Counter(filtered_words)
Shivaji_freq = term_freq['Shivaji']
print("\nTerm Frequency of 'Shivaji':")
print(Shivaji_freq)
```

### Frequency Distribution

```python
import matplotlib.pyplot as plt
freqDist = nltk.FreqDist(word_tokenize(story))
freqDist.plot(10)
```

### Token Count

```python
from nltk.tokenize.regexp import WhitespaceTokenizer
tokens = WhitespaceTokenizer().tokenize(story)
print(len(tokens))
```

### Sentence Count

```python
total_sentences = len(sentences)
print("\nTotal Number of Sentences:")
print(total_sentences)
```

### Stopword Frequency

```python
stopword_freq = Counter(removed_stopwords)
most_common_stopwords = stopword_freq.most_common(3)
print("\nMost Frequent Stopwords:")
print(most_common_stopwords)
```

### Longest Word

```python
longest_word = max(filtered_words, key=len)
print("\nLongest Word:")
print(longest_word, len(longest_word))
```

### Word Cloud

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_words))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('on')
plt.show()
```

### Noun Identification

```python
pos_tagged = nltk.pos_tag(words)
nouns = [word for word, pos in pos_tagged if pos.startswith('NN')]
print("\nNouns in the Story:")
print(nouns)
```

### Verb Identification

```python
verbs = [word for word, pos in pos_tagged if pos.startswith('VB')]
print("\nVerbs in the Story:")
print(verbs)
```

### Word Context

```python
context_sentences = [sentence for sentence in sentences if 'empire' in sentence.lower()]
print("\nSentences with 'empire':")
print(context_sentences)
```

### Named Entity Categorization

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(story)
people = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
locations = [ent.text for ent in doc.ents if ent.label_ == 'GPE']
organizations = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
print("\nNamed Entity Categorization:")
print("People:", people)
print("Locations:", locations)
print("Organizations:", organizations)
```
